# How to Change Plot Features? 
***
> ## The following are done with inputs to your plot method:
>> * <b>Linewidth
>> * Linestyle
>> * Markersize
>> * alpha (_is opacity_)
>> * _color is specified with the letter in `''`_</b>

```python
x=[1,2,3,4,5,6,7]
y=[50,60,65,49,54,65,60]
plt.plot(x,y,'g', linewidth=5, linestyle='dashdot', markersize=105, alpha=1)
plt.xlabel('Date')
plt.ylabel('Temp')
plt.title('Weather in SF')
plt.style.use("seaborn")

#TIP: a shorter way to specifiying the linestyle is to put a simple like '--' next to the color


```

# How Can I Plot More than One Line on a Plot? 
***
> ## Simply make different dependent variable for your independent.

> <b>The following features are demonstrated:
>> * Legend (_Notice how each line needs to be give a class `label`_)
>> * Location,shadow, and fontsize of legend 
>> * Grid (_boolean_)</b>


```python
x=[1,2,3,4,5,6,7]
w=[12,13,34,23,12,11,5]
y=[20,11,5,4,12,21,15]
z=[4,34,2,23,12,34,5]
plt.plot(x,y,color='orange')
plt.plot(x,z, color='red',label='small')
plt.plot(x,w, color='green', label='crazy')
plt.xlabel('Date')
plt.ylabel('Temp')
plt.title('Weather in SF')
plt.legend(loc='lower right', fontsize='large',shadow=True)
plt.grid()
plt.style.use('classic')
```

# Let's do some discrete data: <font color='red'>BarCharts</font>
___


## Here is a random set of data to illustrate the point:
> ### Display the data as lists and let's use NUmpy to represent it!
  
> ### `len` will display the numebr of elements.

```python
company=['Google', 'Amazon', 'Uber', 'Apple']
revenue=[40,50,30,60]
xpos=np.arange(len(company))
xpos
```

# <font color='red'>Then put in your variables and plot it!!</font>

```python
plt.bar(company,revenue)
```

# <font color='red'>Now lets use some of the features changes we learned earlier!</font>

```python
plt.bar(xpos,revenue, label='Revenue')
plt.xticks(xpos,company)    #ticks puts the names on the number values 
plt.title('US Tech Giant Stocks')
plt.ylabel('revenue(billions)')
plt.legend()
```

# What if I Want to Compare more than One Variable? 
***
> ## Let's define another variable `profit` and add it into the previous plot.

```python
profit=[10,15,5,25]
ypos=np.arange(len(profit))
ypos
```

```python
plt.bar(xpos,revenue, label='Revenue')
plt.bar(xpos,profit, label='Profit') #this is to place the profit bars over the revenue bars 
plt.xticks(ypos,company)
plt.title('US Tech Giant Stocks')
plt.ylabel('revenue(billions)')
plt.legend()
```

# What if I Don't Want Overlapping Bars?
***
> ### Simply add in `-/+ pts` by however far you want to sperate them.

> ### Changing the width of the bar also helps seperate them more for clarity.

> ### <font color='red'>Notice how we left out `plt.xticks` and it's now just numeric values</font>

```python
profit=[10,15,5,25]
plt.bar(xpos-.2,revenue, label='Revenue')#.2 moves the positions of the bars
plt.bar(xpos+.2,profit, label='Profit', width=.5)
plt.title('US Tech Giant Stocks')
plt.ylabel('revenue(billions)')
plt.legend()
```

# What if I Want to Change the Orientation of my Plot?
***
> ## Simply add an `h` to the `plt.bar` for horizontal.

```python
profit=[10,15,5,25]
plt.barh(xpos-.2,revenue, label='Revenue')
plt.barh(xpos+.2,profit, label='Profit')
plt.yticks(ypos,company)   #change yticks to make y axis have the companies now 
plt.title('US Tech Giant Stocks')
plt.ylabel('Companies')
plt.xlabel('Revenue(Billions)')
plt.legend()
```

# Ploting Multiple Plots Together: <font color='orange'>Subplots</font>
***
> ## The Subplot function allows you to plot multiple plots all in one figure.  
  
  
  >> ## The concepts of <mark>figure, axes, and plots</mark> tend to be a little confusing in Matplotlib. For example, 'Axes' doesn't mean the cartensian planes as you normally think of it, but means the actual 'plot.'
  >>> ### Take a look at the image below for clarity:

<center><img class="aligncenter wp-image-2087 size-full" src="https://files.realpython.com/media/fig_map.bc8c7cabd823.png"></center>

<center>Image by realpython.com</center>



# <font color='red'>Let's see how we can seperate a single plot into to two plots.</font>

```python
plt.style.use('seaborn') 

data = pd.read_csv('data.csv')
ages = data['Age']
doc_salaries = data['Doctors']
ly_salaries = data['Lawyers']
pb_salaries = data['Pizza Boys']

#By making 2 figures, we have specified there will be two seperate visuals 
#any plot labeled as ax1 will belong to fig1 and ax2 to fig2

fig1, ax1 = plt.subplots()     #no numbers here defualts to a 1x1 axes 
fig2, ax2 = plt.subplots()
#each fig is a dif axes and each axes has plots

ax1.plot(ages, doc_salaries, color='#444444',
         linestyle='--', label='Docs')

ax2.plot(ages, doc_salaries, color='#444444',
         linestyle='--', label='Doctors')
ax2.plot(ages, ly_salaries, label='Lawyers')
ax2.plot(ages, pb_salaries, label='Pizza Boys')
#the second axes(aka the second figure) has 3 plots, see ax2.plot, there are 3 of them

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')


ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()

```

## <font color='red'>Let's now play with it more and take out the plot for Doctors completely out of fig2.</font> 

```python
plt.style.use('seaborn') 

data = pd.read_csv('data.csv')
ages = data['Age']
doc_salaries = data['Doctors']
ly_salaries = data['Lawyers']
pb_salaries = data['Pizza Boys']

fig1, ax1 = plt.subplots() 
fig2, ax2 = plt.subplots()

ax1.plot(ages, doc_salaries, color='#444444',
         linestyle='--', label='Docs')

ax2.plot(ages, ly_salaries, label='Lawyers')
ax2.plot(ages, pb_salaries, label='Pizza Boys')
#the second axes(aka the second figure) has 2 plots

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
#dont know what this does
plt.show()
```

# <font color='red'>Now for the finale, let's split it up into 3 different plots; one for each line. 
  >> ## For certain types of data, it might be better to have it displayed on all seperate figures. 
  >> # The following demonstrates:
  >>> ## This time in the subplot function, we need to specify the RxC to get 3 different figures, if not specified as it wasn't before, it defualts to 1x1

```python
data = pd.read_csv('data.csv')
ages = data['Age']
doc_salaries = data['Doctors']
ly_salaries = data['Lawyers']
pb_salaries = data['Pizza Boys']

fig, (ax1,ax2,ax3) = plt.subplots(nrows=3,ncols=1,sharex=True) 
#if i swtich to rows=1,cols=2, it will print 2 plots side by side
#sharex made the ticks of x shared bw both 

ax1.plot(ages, doc_salaries, color='#444444',
         linestyle='--', label='Docs')
ax2.plot(ages, ly_salaries, label='Lawyers')
ax3.plot(ages, pb_salaries, label='Pizza Boys')

plt.style.use('seaborn') 

ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')


ax2.set_title('Median Salary (USD) by Age')
ax2.set_ylabel('Median Salary (USD)')

ax3.set_title('Median Salary (USD) by Age')
ax3.set_xlabel('Ages')
ax3.set_ylabel('Median Salary (USD)')



plt.tight_layout()
plt.show()
```
